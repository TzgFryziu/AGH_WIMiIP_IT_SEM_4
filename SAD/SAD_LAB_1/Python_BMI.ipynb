```python
import pandas as pd
```

```python
waga = [50,70,80,90,100,110,120,130,140,150]
wzrost = [180,180,180,180,180,180,180,180,180,180]

dane = pd.DataFrame({"Waga":waga, "Wzrost":wzrost})


dane
```

```python
dane["BMI"] = dane["Waga"] / (dane["Wzrost"]*dane["Wzrost"])*10000
dane
```

```python
kod = []
for BMI in dane['BMI']:
    if BMI < 18.5:
        kod.append('Niedowaga')
    elif BMI < 25:
        kod.append('Waga prawidłowa')
    elif BMI < 40:
        kod.append('Nadwaga')
    else:
        kod.append('Otylosc skrajna')
dane["KOD"] = kod
dane
```
